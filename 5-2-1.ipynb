```python
# datetime : 날짜, 시간
import datetime
```

```python
now = datetime.datetime.now()
```

```python
print(now.year, end=' ')
print(now.month, end=' ')
print (now.day, end=' ')
print (now.hour, end=' ')
print(now.minute, end=' ')
```

```python
# 시간 포맷팅
print(now.strftime('%Y. %m. %d %H: %M:%S' )) # 2000. 01. 01 15:10:12
```

```python
# 시간 포맷팅
print(now.strftime('%Y-%m-%d %H: %M: %S' )) # 2000. 01. 01 15:10:12
```
